**Ejercicio 2**

Escribir un programa en Python que declare tres listas: ‘lista1’, ‘lista2’ y ‘lista3’ de cinco enteros cada uno, pida valores para ‘lista1’ y ‘lista2’ y calcule lista3=lista1+lista2.

```python
lista1 = []
lista2 = []
lista3 = []

# Ciclo for para pedir el ingreso de cinco números
for i1 in range(5):
    num1 = int(input('Lista 1: digite un número'))
    lista1.append(num1)

# Ciclo for para pedir el ingreso de cinco números
for i2 in range(5):
    num2 = int(input('Lista 2: digite un número'))
    lista2.append(num2)

# Ciclo for para sumar la relación 1:1 de cada elemento de las listas 1 y 2 y dar los elementos de la lista 3
for i1, i2 in zip(lista1, lista2):
    i3 = int(i1 + i2)
    lista3.append(i3)

print(lista1)
print(lista2)
print(lista3)
```

Escribir un programa en Python que guarde la temperatura mínima y máxima de los
últimos 5 días. El programa debe recibir la información, almacenarla y mostrar:

● La temperatura media de cada día

● Los días con menor temperatura

● Después de almacenar la información de los 5 días, el programa debe recibir una temperatura más por teclado y mostrar los días cuya temperatura máxima coincide con ella. si no existe ningún día se muestra un mensaje de información.

```python
# Diccionario para almacenar las temperaturas mínimas y máximas de cada día
temperaturas = {}

# Input de temperaturas de los últimos 5 días
for d in range(1, 6):
    temp_min = int(input(f'Digite la temperatura mínima del día {d}'))
    temp_max = int(input(f'Digite la temperatura máxima del día {d}'))
    temperaturas[d] = {'min': temp_min, 'max': temp_max}

print(temperaturas)

# Calcular la temperatura media de cada día
for dia, temp in temperaturas.items():
    temp_media = (temp['min'] + temp['max']) / 2
    print(f'La temperatura media del día {dia} fue de {temp_media:.1f} grados centígrados')

# Encontrar los días con menor temperatura
temp_mini_gral = []
for temp in temperaturas.values():
    min = temp['min']
    temp_mini_gral = min
# print(temp_mini_gral)
dias_menor_temp = []
for d, temp in temperaturas.items():
    if temp['min'] == temp_mini_gral:
        dias_menor_temp.append(d)
print(f'La menor temperatura registrada fue de {temp_mini_gral} grados centígrados presente el los días {dias_menor_temp}')

# Ingresar una temperatura adicional y verificar si coincide con la temperatura máxima de alguno de los 5 días anteriores
nueva_temp = int(input('Digita la temperatura del día 6'))
dias_coincidencia = [d for d, temp in temperaturas.items() if temp['max'] == nueva_temp] # Se usa técnica de listas por compresión usando como referencia https://www.youtube.com/watch?v=dkAjT3sDt4I&list=PLh7JzoyIyU4LF1LBB6l-wfouC9Q2Yc_I0
print(f'La temperatura registrada para el día 6 fue de {nueva_temp} grados')
if dias_coincidencia:
    print(f'Los días en los que la temperatura máxima fue {nueva_temp} grados son los días: {dias_coincidencia}')
else:
    print(f'No hay días con una temperatura máxima de {nueva_temp} grados')
```

**Ejercicio 4**

Crea un programa en Python que permita adivinar un número. La aplicación genera un número aleatorio del 1 al 100. A continuación, va pidiendo números y va respondiendo si el número a adivinar es mayor o menor que el introducido, así como el número de intentos que quedan, se contarán con 10 intentos para adivinar el número). El programa termina cuando se acierta el número y debe indicar en que intento fue acertado, si se llega al límite de intentos, el programa debe mostrar que se había generado.

```python
# Generar un número aleatorio entre 1 y 100 con función random
import random
num_random = [random.randint(1,100) for i in range(1)]
# print(num_random)

# Guardar los números ingresados en cada iteración for y compararlos con el número aleatorio por medio de control if.
# En el print de las iteraciones, se avisa los intentos que quedan restando al rango el índice del número que se ingresa.
num_ingresados = []
for num in range(10):
    num = int(input('Adivine un número del 1 al 100'))
    num_ingresados.append(num)
    if num < num_random[0]:
        print(f'El número es mayor a {num}. Te quedan {10-(num_ingresados.index(num)+1)} intentos')
    elif num > num_random[0]:
        print(f'El número es menor a {num}, Te quedan {10-(num_ingresados.index(num)+1)} intentos')
    else:
        num == num_random[0]
        print(f'Intento {num_ingresados.index(num)+1} ¡Adivinaste! el número es el {num_random[0]}')
        break
# print(num_ingresados)
```
